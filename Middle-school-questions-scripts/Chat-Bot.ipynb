```python
import pandas as pd
import numpy as np
```

```python
sample=pd.read_csv("final_key.csv")
```

```python
sample.head()
```

```python
ans=pd.read_json("final_key.jsonl",lines=True)
```

```python
stem=[]
ls={"A":[],"B":[],"C":[],"D":[],"E":[]}
ls.keys()
```

```python
for i in range(len(ans)):
    question=pd.DataFrame(ans.loc[i]["question"])
    tt=question["choices"]
    labels=[k['label'] for k in tt]
    for idx,key in enumerate(labels):
        ls[key].append(tt[idx]["text"])
    for f in ["A","B","C","D","E"]:
        if f not in labels:
            ls[f].append("")
    stem.append(question["stem"].loc[0])
    
    
```

```python
for i in ["A","B","C","D","E"]:
    ans[i]=ls[i]
ans["stem"]=stem    
    
```

```python
ans.head()
```

```python
ans=ans.drop(["question"],axis=1)
ans=ans.drop(["E"],axis=1)
```
